```python
import pandas as pd
```

###  Drop the Null Values using Dropna()

```python
df = pd.read_csv('C:\\Users\\abhishek\\Downloads\\Fortune_10.csv')
df
```

```python
# remove Null Nan values 
df.dropna()
```

#### Drop Null values Column

```python
df.dropna(axis=1)
```

```python
df.dropna(how='any')
```

```python
df.dropna(axis = 1, how = 'any')
```

```python
df.dropna(how='all')
```

```python
# minimum 1 not null value 
df.dropna(thresh=1)
```

```python
df.dropna(thresh=5)
```

```python
df = df.dropna(subset=['Name'])
df
```

```python
df.dropna(inplace=True)
df
```
