## Pandas Merging Function ()

```python
import pandas as pd
```

#### Pandas Merge funtion conects columns or indexes in Dataframe based on one or more keys.

```python
import pandas as pd
```

### Creating DataFrames

```python
df1 = pd.DataFrame({'ID' :[1,2,3,4], 
                  'Class': [9,10,11,12]})
df1
```

```python
df2 = pd.DataFrame({'ID': [1,2,3,5],
                   'Name':['A','B','C','D']})
df2
```

### Merge

```python
# on parameter
pd.merge(df1,df2, on = 'ID')
```

### Changing Position of Coloumn (left,right parameter)

```python
pd.merge(df2,df1, on = 'ID')
```

### If coloumn(ID) is not similar

```python
pd.merge(df1, df2, on ='ID')
```

### Inner Merge ( common in both database)

```python

pd.merge(df1, df2, on ='ID' , how ='inner')
```

### Left Merge

```python
pd.merge(df1, df2, on ='ID' , how ='left')
```

### Right Merge

```python
pd.merge(df1, df2, on ='ID' , how ='right')
```

### Outer Merge (all different values)

```python
pd.merge(df1, df2, on ='ID' , how ='outer') 
```

### Outer Merge indicator

```python
pd.merge(df1, df2, on ='ID' , how ='outer',indicator = True)   #left=df1 #right=df2
```

### If Both Dataframe are Same

```python
df1 = pd.DataFrame({'ID' :[1,2,3,4], 
                  'Class': [9,10,11,12]})
df1
```

```python
df2 = pd.DataFrame({'ID' :[1,2,3,4], 
                  'Class': [9,10,11,12]})
df2
```

### Suffixes

```python
pd.merge(df1,df2, on = 'ID', suffixes =('_Higher','_Middle'))
```
