### Pandas fillna()

```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\abhishek\\Downloads\\Fortune_10(1).csv')
df
```

```python
# Nan is replacedy by 0
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'Name': 'none', 'Profit':0,'Growth':0})
```

### Method Forward Fill ('ffill')
 It print previous value

```python
df.fillna(method = 'ffill')
```

```python
df.fillna(method = 'pad')     ## ffill = pad
```

### Backward Fill('bfill')

```python
df.fillna(method = 'bfill')
```

### Axis

```python
df.fillna(method = 'ffill',axis = 0)
```

```python
df.fillna(method = 'bfill',axis = 0)
```

```python
df.fillna(method = 'bfill', axis = 1)
```

```python
df.fillna(0, limit = 1)
# zero only at 1 place in coloumn
```

```python
df.fillna(0, limit = 2)
```

```python
df.fillna(method ='ffill', limit = 2)
```

### Filna Inplace 

```python
df.fillna(5, inplace = True)
df
```
