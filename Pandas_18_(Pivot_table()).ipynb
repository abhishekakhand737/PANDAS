## Pivot Table in Pandas

##### A pivot table stores multiple levels using a MultiIndex in the rows and columns of the result DataFrame.

```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\abhishek\\Downloads\\movies_data.csv')
df
```

### Pivot_table Function () with different Parameters

### Index parameter 

```python
#Agrecated or mean Value
df.pivot_table(index = 'Genre')
```

### Column parameter

```python
#Mean Value (YEAR wise information)
df.pivot_table(index = 'Genre', columns = 'Year')    
```

### Aggregation function 

```python
df.pivot_table(index = 'Genre', columns = 'Year' , aggfunc = 'count')    
```

### Max Aggregation Function

```python
df.pivot_table(index = 'Genre', columns = 'Year' , aggfunc = 'max')    
```

### Sum

```python
df.pivot_table(index = 'Genre', columns = 'Year' , aggfunc = 'sum')    
```

### Fill Value Parameters

```python
df.pivot_table(index = 'Genre', columns = 'Year' , fill_value = 'none')    
 # nan value replaced by none
```

```python
df.pivot_table(index = 'Genre', columns = 'Year' , fill_value = '0')
```

### Margin  (Grand total values)

```python
df.pivot_table(index = 'Genre', columns = 'Year' , fill_value = 0, margins = True)
```

```python
df.pivot_table(index = 'Genre', columns = 'Year' , aggfunc ='sum', fill_value = 0, margins = True)
```
