## Loc & iLoc

loc is used to select data by row and column labels (names).   
iloc is used to select data by integer position (row number & column number).    

```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\abhishek\\Downloads\\student_social_study.csv')
df
```

## Loc (Label Based)

```python
df.loc[0]
```

```python
df.iloc[4]
```

### Lists or Array of labels

```python
df.loc[[0,1]]
```

```python
df.loc[[2,4]]
```

```python
int   (df.loc[4,'Class'] )
```

```python
df.loc[0:3,'Class']
```

```python
df.loc[0:2,'Percentage']
```

```python
df.loc[[ False,  False, True, False, False ]]
```

```python
df.loc[df['Class'] < 11,['Percentage']]
```

## iLoc (Index based)

```python
df.iloc[0]
```

```python
df.iloc[[0]]
```

```python
df.iloc[[0,1]]
```

```python
df.iloc[[True, False, True, False,False]]
```
