## Pandas Series
Pandas Series is a one-dimensional labeled array in Python 
that can hold data of any type such as integers, floats, strings, etc.

```python
import pandas as pd
```

```python
pd.__version__
```

### Create a list with different data types

```python
list_s = [1, 2, -3, 6.5, 'data values']
print(list_s)
```

### Convert List into Series PANDAS

```python
# 1st column index value  2nd column data values.
series1 = pd.Series(list_s)
print(series1)
```

### Data type of Pandas Series

```python
type(series1)
```

### Create a Series

```python
series2 = pd.Series([1,2,3,4])
print(series2)
```

```python
### Create a Empty Series
empty_s = pd.Series([])
print(empty_s)
```

### Creating a Pandas Series with Custom Index.

series3 = pd.Series([1,2,3,4], index = ['a','b','c','d'])
print(series3)

### Changing Data Type

series3 = pd.Series([1,2,3,4], index = ['a','b','c','d'], dtype = float, name = 'data values')
print(series3)

### Create Pandas Series using scalar Value

scalar_s = pd.Series(0.8)
print(scalar_s)

```python
scalar_s = pd.Series(0.8, index = [1,2,3])
print(scalar_s)
```

### Creating Series from Dictionary

```python
dict_s = pd.Series({'a':1,'b':2})
print(dict_s)
```

### Creating a Pandas Series with Default Index

```python
s4 = pd.Series([1,2,3,4,5])
print(s4)
```

```python
int(  s4[0]                   )
```

```python
s4[3]
```

### Slicing a Pandas Series

```python
s4[0:3]
```

#### Maximum Value of Series

```python
int(      max(s4)  )
```

#### Minimum Value of Series

```python
int(     min(s4)  )
```

```python
s4[s4 > 3]
```

```python
# Both series having equal value.
s4 + s5
```
