## Pandas Join Function ()

The join() function in Pandas is used to combine two DataFrames based on their index or a key column.

```python
import pandas as pd
```

### Creating Two Dataframe

```python
df1 = pd.DataFrame({'A': [1,2,3],
                    'B':[10,20,30]})
df2 = pd.DataFrame({'C': [4,5,6],
                    'D':[40,50,60]})
display(df1,df2)
```

```python
df1.join(df2)
```

## Different Index Dataframe

```python
df1 = pd.DataFrame({'A': [1,2,3],
                    'B':[10,20,30] }, index = ['a','b','c'])
df2 = pd.DataFrame({'C': [4,5,6],
                    'D':[40,50,60]})
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1 = pd.DataFrame({'A': [1,2,3],
                    'B':[10,20,30] }, index = ['a','b','c'])
df2 = pd.DataFrame({'C': [4,5],
                    'D':[40,50] },  index = ['a','b'])
display(df1,df2)
```

```python
df1.join(df2)
```

### Right Join

```python
df1.join(df2, how = 'right')
```

### Outer Join

```python
df1.join(df2, how = 'outer')
```

### Same Column Name

```python
df1 = pd.DataFrame({'A': [1,2,3],
                    'B':[10,20,30] }, index = ['a','b','c'])
df2 = pd.DataFrame({'A': [4,5],
                    'D':[40,50] },  index = ['a','b'])
display(df1,df2)
```

```python
df1.join(df2, how='outer', lsuffix='_left', rsuffix='_right')
```
