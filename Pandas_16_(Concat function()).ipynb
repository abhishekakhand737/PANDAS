## Pandas Concat Function()

```python
import pandas as pd
```

### Creating a Series

```python
sr1 = pd.Series([0,1,2])
sr1
```

```python
sr2 = pd.Series([3,4,5])
sr2
```

### Concat a Series 

```python
pd.concat([sr1,sr2])
```

### Concat enqual value Series

```python
sr1 = pd.Series([0,1,2])
sr1
```

```python
sr2 = pd.Series([3,4,5,6,7])
sr2
```

```python
pd.concat([sr1,sr2])
```

### Concat DataFrame

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'Name':['A','B','C','D'],
             'Class': [5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID':[5,6,7,8],
             'Name':['E','F','G','H'],
             'Class': [9,10,11,12]})
df2
```

```python
pd.concat([df1,df2])
```

### Changing Position of Dataframe

```python
pd.concat([df2,df1])
```

### Concat Columns Wise (axis=1)

```python
pd.concat([df2,df1], axis = 1)
```

### Concat Rows Wise (axis=0)

```python
pd.concat([df2,df1], axis = 0)
```

```python
pd.concat([df1,df2], axis = 0, ignore_index = True)
```

## Unequal Dataframe Concat

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'Name':['A','B','C','D'],
             'Class': [5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID':[3,4],
             'Name':['C','D'],
             'Class': [7,8]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis = 1)
```

### pd.concat() with axis=1 and inner join

```python
pd.concat([df1,df2], axis = 1,join = 'inner')
```

### Pandas Concat Function and Join Operations

```python
# pd.concat([df1, df2], axis = 1, join_axes = [df2.index])
```

```python
pd.concat([df1, df2], axis=1).reindex(df2.index)
```

### keys Parameters (Multiple Dataframe Separated)

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'Name':['A','B','C','D'],
             'Class': [5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID':[5,6,7,8],
             'Name':['E','F','G','H'],
             'Class': [9,10,11,12]})
df2
```

```python
pd.concat([df1, df2], keys = ['df1','df2'])
```

```python
pd.concat([df1, df2], keys = ['First df1','Second df2'])
```

```python
pd.concat([df1, df2], axis =1, keys = ['First df1','Second df2'])
```

### Different Column Names

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'Name':['A','B','C','D'],
             'Class': [5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'Marks': [40,60,80,100]})
df2
```

```python
pd.concat([df1,df2])
```
